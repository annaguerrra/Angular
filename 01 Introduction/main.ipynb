# Aula 01 Introduction 

## **Conceitos importantes**

### 1. SPA (Single Page Application)

É um tipo de aplicação web que carrega apenas uma página HTML e atualiza o conteúdo dinamicamente sem recarregar a página inteira quando o usuário navega. O navegador carrega a aplicação uma única vez. Depois disso, as mudanças de tela acontecem através de JavaScript, que altera partes da página.

O Angular funciona como um framework front-end que controla toda a interface da aplicação no navegador.

### 2. Router
O Angular Router controla a navegação entre telas.

``` typescript
const routes: Routes = [
  { path: 'home', component: HomeComponent },
  { path: 'produtos', component: ProdutosComponent },
  { path: 'contato', component: ContatoComponent }
];    
```

Quando o usuário acessa `/produtos`, o Angular troca apenas o componente, sem recarregar a página.

### 3. Componentes

Cada "tela" da aplicação é um componente Angular. Um componente standalone é um componente independente, que pode ser usado diretamente na aplicação sem ser declarado em um módulo (NgModule).

**x.component.css:** O css que e aplicado apenas a este componente  
**x.component.html:** o html a ser reporduzido ao utilizarmos este componente  
**x.component.spec.ts:** Especificação de testes do componente  
**x.component.ts:** Classe TypeScript que define comportamento do componente.   


### 4. Binding
Permite conectar eventos e informações ao código de comportamento do componente.

## Ordem dos Arquivos
SRC  
  |- APP  
     |- Features (pages)  
        |- LoginPage  
        |- MainPage  
        |- NotFoundPage  
     |- Shared  
        |- Header  
        |- Button  
        |- Input
